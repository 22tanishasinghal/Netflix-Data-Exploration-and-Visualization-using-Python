**Analysis for Director**

```python
data_agg = df.groupby(by ='director')[['year', 'title']].aggregate({"year": ['min', 'max'], "title": 'count'})
data_agg
```

```python
data_agg.columns = ["_".join(col) for col in data_agg.columns]
data_agg
```

```python
data_agg['yrs_active'] = data_agg['year_max']  - data_agg['year_min']
data_agg.sort_values(by=['title_count','yrs_active'], ascending=False)
```

Insights:
- Maximum number of movies or TV shows are directed by 'Cathy Garcia-Molina' in only 11 years.
- The least number of movies or TV shows are directed by 'Yoo Byung-jae'. It has not been a complete one year since he started directing and he has directed 1 till now.



**Most Active Director**

```python
data_agg["type_per_yr"] = data_agg["title_count"] / data_agg["yrs_active"]
data_agg
```

Since for some directors, number of active years is Zero: there types per year is coming out to be INFINITY.
Let's filter those out

```python
data_agg.loc[data_agg['yrs_active'] > 0].sort_values(by='type_per_yr',ascending=False)
```

Insights:
- Most active director is Cathy Garcia-Molina.
- Least active director is Shalini Kantayya
